**Handling Outlier**

```python
sns.boxplot(x='SepalWidthCm', data=iris)
plt.figure(figsize=(0.1,0.1))
```



*   The values above 4 and below 2.25 are acting as outliers.




**Remove the outliers**

```python
iris.shape
```

```python
# Inter Quartile Range
Q1 = np.percentile(iris['SepalWidthCm'], 25,
                interpolation = 'midpoint')
 
Q3 = np.percentile(iris['SepalWidthCm'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
 
# Upper bound
upper = np.where(iris['SepalWidthCm'] >= (Q3+1.5*IQR))
 
# Lower bound
lower = np.where(iris['SepalWidthCm'] <= (Q1-1.5*IQR))
 
# Removing the Outliers
iris.drop(upper[0], inplace = True)
iris.drop(lower[0], inplace = True)

sns.boxplot(x='SepalWidthCm', data=iris)
print("After removing the outliers the dataset size is :",iris.shape)
```

**Model building**

```python
from sklearn.model_selection import train_test_split

train,test=train_test_split(iris,test_size=0.25)
```

```python
X_train = train[['SepalLengthCm',"SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
y_train = train.Species
```

```python
X_test = test[['SepalLengthCm',"SepalWidthCm","PetalLengthCm","PetalWidthCm"]]
y_test = test.Species
```

**Importing Decision Tree from Sklearn to build a classification model**

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
dt = DecisionTreeClassifier(random_state=12)
model = dt.fit(X_train, y_train)
```

```python
dt.predict(X_test)
```

```python
# Check accuracy of model
dt.score(X_test, y_test)
```

```python
#prediction
y_pred = dt.predict(X_test)
```

```python
#Accuracy
from sklearn import metrics
print('Accuracy Score:', metrics.accuracy_score(y_test, y_pred))
```
